```python
import sys

# 구간 N개 테스트 M번
N,M = map(int, sys.stdin.readline().strip().split())

# 속도 제한 구간
Lsection = []

# 제한 속도
Llimit= []

# 시작 구간
start= 0

for _ in range(N):
    sec, lim = map(int, sys.stdin.readline().strip().split())
    start+= sec
    Lsection.append(start)
    Llimit.append(lim)

#테스트 구간
Tsection = []

#테스트 속도
Tspeed = []
#시작구간
start = 0

for _ in range(M):
    sec,speed = map(int, sys.stdin.readline().strip().split())
    start += sec
    Tsection.append(start)
    Tspeed.append(speed)

# 테스트 구간과 제한속도 구간을 고려해 구간을 나누어 보기
section = []

for i in Lsection:
    section.append(i)
for i in Tsection:
    if i not in section:
        section.append(i)
section.sort()

# 해당 구간의 제한속도 구하는 함수
def limitCheck(sec):
    for Ls, Ll in zip(Lsection,Llimit):
        if sec <= Ls:
            return Ll

# 해당 구간의 테스트 속도를 구하는 함수
def speedCheck(sec):
    for Tsec, Tsp in zip(Tsection,Tspeed):
        if sec <= Tsec:
            return Tsp

# 문제의 정답
exceed_max = 0

#구간마다 체크하기
for i in section:
    # 구간의 제한속도 구하기
    limitSpeed = limitCheck(i)
    # 구간의 실제 테스트 속도 구하기
    testSpeed = speedCheck(i)
    
    # 제한 속도를 초과했을 경우
    if testSpeed> limitSpeed:
        # 속도차 exceed
        exceed = testSpeed-limitSpeed
        # 저장된 exceed_max 보다 현재 exceed가 크다면 exceed_max값을 현재 exceed값으로 대체
        if exceed > exceed_max:
            exceed_max = exceed
            



print (exceed_max)
```
